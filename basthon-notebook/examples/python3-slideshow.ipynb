```python
def eratosthenes(nmax):
    """ Retourne la liste des nombres premiers inférieurs ou égaux à nmax > 1."""
    table = set(range(2, nmax + 1))
    premiers = []
    while table:
        p = min(table)
        table -= set(range(p, nmax + 1, p))
        premiers.append(p)
 
    return premiers
```

```python
eratosthenes(100)
```
