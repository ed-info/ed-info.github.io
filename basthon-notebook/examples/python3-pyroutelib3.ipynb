```python
from pyroutelib3 import Router
import folium

depart = [47.08428480854615, 2.3939454203269217]
arrivee = [47.08187952227819, 2.398992969262373]
milieu = [sum(c) / 2 for c in zip(depart, arrivee)]

m = folium.Map(location=milieu, zoom_start=17)

folium.Marker(depart, popup="Départ").add_to(m)
folium.Marker(arrivee, popup="Arrivée").add_to(m)

router = Router("foot")
node_depart = router.findNode(*depart)
node_arrivee = router.findNode(*arrivee)
status, route = router.doRoute(node_depart, node_arrivee)
if status == 'success':
    routeLatLons = list(map(router.nodeLatLon, route))
    folium.PolyLine(routeLatLons, color="green", weight=5, opacity=0.7).add_to(m)

m
```
