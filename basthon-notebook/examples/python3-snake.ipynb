```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display_image
```

```python
def read(filename):
    """ Read an image from a file into an array. """
    return (255 * plt.imread(filename)).astype(int)

def show(*args):
    """ Display numpy arrays as images. """
    for img in args:
        display_image(img)
```

```python
snake = read("snake.png")
show(snake)
```

```python
from scipy.ndimage import gaussian_filter

# see https://en.wikipedia.org/wiki/Unsharp_masking
blurred = gaussian_filter(snake, sigma=2)
sharp = np.clip(snake + (snake - blurred), 0, 255)

show(snake, sharp)
```

```python
from scipy.ndimage import sobel

show(snake, sobel(snake))
```

```python
from scipy.ndimage import median_filter

show(snake, median_filter(snake, 7))
```
