```python
import graphviz
import basthon
```

```python
dot = """digraph g {

node [shape = circle,
      style = filled,
      color = grey]

node [fillcolor = red]
a

node [fillcolor = green]
b c d

node [fillcolor = orange]

edge [color = grey]
a -> {b c d}
b -> {e f g h i j}
c -> {k l m n o p}
d -> {q r s t u v}
}"""
```

```python
graphviz.Source(dot)
```

```python
for engine in ('dot', 'neato', 'twopi', 'circo'):
    basthon.display(graphviz.Source(dot, engine=engine))
```
