```ocaml
let read_file filename = 
  let ch = open_in filename in
    let s = really_input_string ch (in_channel_length ch) in
      close_in ch;
      s
```

```ocaml
let reverse s =
  let len = String.length s in
    String.init len (fun i -> s.[len - 1 - i])
```
