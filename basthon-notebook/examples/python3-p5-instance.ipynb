```python
from p5.instance import *
```

```python
def setup(sketch):
    sketch.createCanvas(200, 200)
    sketch.background(160)

def draw(sketch):
    sketch.fill("blue")
    sketch.background(200)
    radius = sketch.sin(sketch.frameCount / 60) * 50 + 50
    sketch.ellipse(100, 100, radius, radius)

run(setup, draw)
```

```python
def draw(sketch):
    sketch.fill("red")
    sketch.background(200)
    radius = sketch.sin(sketch.frameCount / 60) * 50 + 50
    sketch.ellipse(100, 100, radius, radius)

run(setup, draw)
```
