```python
import turtle


def recursive_drawing_nested(func):
    def draw_func(length, depth):
        def forward(factor):
            draw_func(length * factor, depth - 1)
        if depth == 0:
            turtle.forward(length)
        else:
            func(forward)
    draw_func.mode = "nested"
    return draw_func


def recursive_drawing_extend(func):
    def draw_func(length, depth):
        def forward(factor):
            turtle.forward(length)
            draw_func(length * factor, depth - 1)
            turtle.penup()
            turtle.backward(length)
            turtle.pendown()
        if depth > 0:
            func(forward)
    draw_func.mode = "extend"
    return draw_func


def recursive_drawing(mode="nested"):
    if mode == "nested":
        return recursive_drawing_nested
    elif mode == "extend":
        return recursive_drawing_extend


def draw(length, depth, recurssive_func):
    turtle.shape("turtle")
    turtle.color("red")
    turtle.speed("fastest")

    if recurssive_func.mode == "nested":
        turtle.penup()
        turtle.backward(length / 2)
        turtle.pendown()

    recurssive_func(length, depth)

    turtle.done()
```

```python
@recursive_drawing(mode="nested")
def von_koch(forward):
    forward(1/3)
    turtle.left(60)
    forward(1/3)
    turtle.right(120)
    forward(1/3)
    turtle.left(60)
    forward(1/3)

draw(600, 3, von_koch)
```

```python
@recursive_drawing(mode="extend")
def tree(forward):
    turtle.left(30)
    forward(0.6)
    turtle.right(60)
    forward(0.6)
    turtle.left(30)

# tronc
turtle.color("red")
turtle.penup()
turtle.left(90)
turtle.backward(150)
turtle.pendown()
turtle.forward(150)

draw(100, 8, tree)
```
