```ocaml
let random_char () =
  (Char.chr (97 + (Random.int 26)))
```

```ocaml
let rec random_string ?(acc="") length =
  match length with
  | 0 -> acc;
  | _ -> random_string ~acc:((String.make 1 (random_char ())) ^ acc) (length - 1)
```

```ocaml
let txt = random_string 10000 in
  let oc = open_out "random.txt" in 
    output_string oc txt;
    close_out oc 
```
