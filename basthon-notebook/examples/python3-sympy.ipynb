```python
from sympy import *
```

```python
x = Symbol("x")
sol = solve(x ** 2 + x + 1)
pretty_print(sol)
```

```python
area_gaussian = Integral(exp(-x ** 2), (x, -oo, oo))
pretty_print(area_gaussian, "=", area_gaussian.doit())
```

```python
n = Symbol("n")
pretty_print(summation(1 / n ** 2, (n, 1, oo)))
```
